```python
# Problema 8
def potencia(base, exponente):
  if exponente==0:
    return 1
  elif exponente>0: 
    return base if exponente==1 else base*potencia(base, exponente-1)
  else:
    return 1/base if exponente==-1 else (1/base)*potencia(base, exponente+1)
 
def factorial(n):
  return 1 if (n==1 or n==0) else n * factorial(n - 1);  

def exponencial(x,n):
  resultado=0.0
  for i in range(n+1):
    resultado+= potencia(x,i)/factorial(i)
  print("El valor de la función exponencial para " + str(x) + " es: "+ str(resultado))

x= float(input("Ingrese l valor de x al que quiere caluar su exponencial: "))
n= int(input("Ingrese el número de términos de la serie de Maclaurin que quiere utilizar: "))    
exponencial(x, n)
```
